Operator aritmatika adalah simbol atau tanda yang digunakan untuk melakukan operasi matematika dasar seperti penjumlahan, pengurangan, perkalian, pembagian, dan lainnya. Berikut adalah beberapa operator aritmatika yang umum:

1. Penjumlahan (+): Menambahkan dua nilai.

Contoh: 3 + 2 = 5



2. Pengurangan (-): Mengurangi satu nilai dengan nilai lainnya.

Contoh: 5 - 2 = 3



3. Perkalian (*): Mengalikan dua nilai.

Contoh: 4 * 3 = 12



4. Pembagian (/): Membagi satu nilai dengan nilai lainnya.

Contoh: 6 / 2 = 3



5. Modulus (%): Menghasilkan sisa dari pembagian dua bilangan.

Contoh: 7 % 2 = 1 (karena 7 dibagi 2 sisanya 1)



6. Pangkat () atau (^)**: Menaikkan bilangan ke pangkat tertentu.

Contoh: 2 ** 3 = 8 (artinya 2 dipangkatkan 3)




Operator ini digunakan dalam berbagai bahasa pemrograman dan juga dalam matematika untuk melakukan perhitungan dasar.
